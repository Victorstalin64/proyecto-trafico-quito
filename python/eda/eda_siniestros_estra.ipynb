```python
import pandas as pd

columnas_mensual = ["provincia", "total", "enero", "febrero", "marzo", "abril",
                     "mayo", "junio", "julio", "agosto", "septiembre", "octubre",
                     "noviembre", "diciembre", "extra"]

df1 = pd.read_csv(
    "../../data/raw/siniestros_2021/1.csv",
    encoding="latin-1",
    skiprows=3,
    names=columnas_mensual,
    thousands=","
)
df1 = df1.drop(columns=["extra"])
df1.head(10)
```

```python
df1_pichincha = df1[df1["provincia"].str.strip() == "Pichincha"]
df1_pichincha
```

```python
print("Nulos por columna:")
print(df1.isnull().sum())
print("\nFilas duplicadas:", df1.duplicated().sum())
```

```python
df1.to_csv("../../data/processed/siniestros_mensual_limpio.csv", index=False)
print("Guardado")
```

```python
df6 = pd.read_csv(
    "../../data/raw/siniestros_2021/6.csv",
    encoding="latin-1",
    skiprows=2,
    header=None,
    thousands=","
)

print("Forma del archivo:", df6.shape)

df6 = df6.iloc[:, :4]
df6.columns = ["provincia", "total_victimas", "fallecidos", "lesionados"]
df6.head(10)
```

```python
df6["provincia"] = df6["provincia"].str.strip()

# Descarta filas de pie de página (notas al final del archivo)
df6 = df6[df6["provincia"].notna()]
df6 = df6[~df6["provincia"].str.contains("Elaboración|Fuente|Datos provisionales", na=False)]

df6_pichincha = df6[df6["provincia"] == "Pichincha"]
df6_pichincha
```

```python
df6.to_csv("../../data/processed/siniestros_victimas_nacional.csv", index=False)
df6_pichincha.to_csv("../../data/processed/siniestros_victimas_pichincha.csv", index=False)
print("Guardado")
```

```python
import codecs
with codecs.open("../../data/raw/estra_2023/4.csv", "r", encoding="latin-1") as f:
    for i, linea in enumerate(f):
        print(linea.strip())
        if i > 8:
            break
```

```python
df_vehiculos = pd.read_csv(
    "../../data/raw/estra_2023/4.csv",
    encoding="latin-1",
    skiprows=3,
    header=None,
    thousands=","
)

print("Forma del archivo:", df_vehiculos.shape)

df_vehiculos = df_vehiculos.iloc[:, :6]
df_vehiculos.columns = ["provincia", "canton", "estado", "alquiler", "particular", "otros"]
df_vehiculos.head(10)
```

```python
df_vehiculos["provincia"] = df_vehiculos["provincia"].str.strip()
df_vehiculos["canton"] = df_vehiculos["canton"].str.strip()
df_vehiculos = df_vehiculos[df_vehiculos["provincia"].notna()]
df_vehiculos = df_vehiculos[~df_vehiculos["provincia"].str.contains("Elaboración|Fuente|Datos provisionales", na=False)]

print("Nulos por columna:")
print(df_vehiculos.isnull().sum())
print("\nFilas duplicadas:", df_vehiculos.duplicated().sum())
```

```python
df_vehiculos_pichincha = df_vehiculos[df_vehiculos["provincia"] == "Pichincha"]
df_vehiculos_pichincha
```

```python
df_vehiculos.to_csv("../../data/processed/vehiculos_matriculados_nacional.csv", index=False)
df_vehiculos_pichincha.to_csv("../../data/processed/vehiculos_matriculados_pichincha.csv", index=False)
print("Guardado")
```
