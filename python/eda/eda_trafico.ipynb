```python
import pandas as pd

df = pd.read_csv("../../data/raw/trafico_quito_completo.csv")
df.shape
```

```python
print("Nulos por columna:")
print(df.isnull().sum())

print("\nFilas duplicadas:", df.duplicated().sum())

print("\nEstadísticas generales:")
df.describe()
```

```python
calidad_antes = {
    "total_filas": len(df),
    "total_columnas": len(df.columns),
    "nulos_totales": df.isnull().sum().sum(),
    "duplicados": df.duplicated().sum(),
}
calidad_antes
```

```python
df_limpio = df.drop_duplicates()

df_limpio = df_limpio[(df_limpio["nivel_congestion"] >= 0) & (df_limpio["nivel_congestion"] <= 100)]

df_limpio = df_limpio[df_limpio["velocidad_promedio_kmh"] >= 0]

df_limpio["fecha_hora"] = pd.to_datetime(df_limpio["fecha_hora"])

df_limpio.shape
```

```python
calidad_despues = {
    "total_filas": len(df_limpio),
    "total_columnas": len(df_limpio.columns),
    "nulos_totales": df_limpio.isnull().sum().sum(),
    "duplicados": df_limpio.duplicated().sum(),
}
calidad_despues
```

```python
print("1. Conteo de registros:", len(df_limpio))

print("2. Congestión fuera de rango:", ((df_limpio["nivel_congestion"] < 0) | (df_limpio["nivel_congestion"] > 100)).sum())

print("3. Combinaciones fecha+zona duplicadas:", df_limpio.duplicated(subset=["fecha_hora", "zona"]).sum())

zonas_validas = df_limpio["zona"].unique()
print("4. Zonas únicas encontradas:", len(zonas_validas))

print("5. Fechas fuera de 2024:", ((df_limpio["fecha_hora"].dt.year != 2024)).sum())
```

```python
df_limpio.to_csv("../../data/processed/trafico_quito_limpio.csv", index=False)
print("Guardado en data/processed/")
```
