```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Semilla para reproducibilidad
np.random.seed(42)

# Zonas de Quito y factores base de densidad vehicular
zonas = [
    "La Carolina", "El Ejido", "La Mariscal", "La Floresta",
    "El Batán", "Cumbayá", "Quitumbe", "Solanda",
    "Iñaquito", "San Rafael"
]

# Rango de fechas: un año completo (2024 bisiesto)
fecha_inicio = datetime(2024, 1, 1)
fecha_fin = datetime(2024, 12, 31)
fechas = pd.date_range(fecha_inicio, fecha_fin, freq="15min")

filas = []

# Climas posibles en Quito para simular impacto en tráfico
climas = ["Despejado", "Lluvia Ligera", "Tormenta", "Neblina"]
prob_clima = [0.65, 0.20, 0.05, 0.10]

for fecha in fechas:
    hora = fecha.hour
    dia_semana = fecha.weekday()  # 0=lunes, 6=domingo
    mes = fecha.month

    # Simulación de clima estocástico diario/horario
    clima_actual = np.random.choice(climas, p=prob_clima)

    # Congestión base según hora pico
    if hora in [7, 8, 17, 18]:
        congestion_base = 80
    elif hora in [6, 9, 16, 19]:
        congestion_base = 60
    elif 22 <= hora or hora <= 5:
        congestion_base = 10
    else:
        congestion_base = 35

    # Menos tráfico el fin de semana
    if dia_semana >= 5:
        congestion_base = congestion_base * 0.5

    for zona in zonas:
        # Variación aleatoria del ruido
        ruido = np.random.normal(0, 8)

        # El clima adverso aumenta la congestión
        factor_clima = 0
        if clima_actual == "Lluvia Ligera":
            factor_clima = 10
        elif clima_actual in ["Tormenta", "Neblina"]:
            factor_clima = 20

        congestion = max(0, min(100, congestion_base + ruido + factor_clima))

        # Simulación de velocidad promedio inversamente proporcional al tráfico (km/h)
        velocidad_promedio = round(max(10, 60 - (congestion * 0.45) + np.random.normal(0, 3)), 1)

        # Tiempo estimado de recorrido por tramo en minutos
        tiempo_recorrido_min = round(5 + (congestion * 0.15) + np.random.uniform(0, 2), 1)

        filas.append({
            "fecha_hora": fecha,
            "mes": mes,
            "hora": hora,
            "dia_semana": dia_semana,
            "es_fin_semana": dia_semana >= 5,
            "zona": zona,
            "clima": clima_actual,
            "nivel_congestion": round(congestion, 1),
            "velocidad_promedio_kmh": velocidad_promedio,
            "tiempo_recorrido_min": tiempo_recorrido_min,
            "vehiculos_estimados": int(congestion * np.random.uniform(15, 25))
        })

df_trafico_completo = pd.DataFrame(filas)
display(df_trafico_completo.head())
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

np.random.seed(42)

zonas = [
    "La Carolina", "El Ejido", "La Mariscal", "La Floresta",
    "El Batán", "Cumbayá", "Quitumbe", "Solanda",
    "Iñaquito", "San Rafael", "Tumbaco", "Conocoto",
    "Chillogallo", "Carapungo", "Calderón", "Guápulo",
    "La Kennedy", "Cotocollao", "El Condado", "Turubamba"
]

tipos_evento = ["Ninguno", "Concierto", "Partido de fútbol", "Feria", "Manifestación", "Evento cultural"]
prob_evento = [0.85, 0.05, 0.03, 0.03, 0.02, 0.02]

fecha_inicio = datetime(2024, 1, 1)
fecha_fin = datetime(2024, 12, 31)
fechas = pd.date_range(fecha_inicio, fecha_fin, freq="5min")

filas = []

for fecha in fechas:
    hora = fecha.hour
    dia_semana = fecha.weekday()

    for zona in zonas:
        tipo_evento = np.random.choice(tipos_evento, p=prob_evento)
        evento_activo = tipo_evento != "Ninguno"

        if evento_activo:
            asistentes_estimados = int(np.random.uniform(200, 15000))
            impacto_trafico = round(np.random.uniform(10, 40), 1)
        else:
            asistentes_estimados = 0
            impacto_trafico = 0.0

        filas.append({
            "fecha_hora": fecha,
            "hora": hora,
            "dia_semana": dia_semana,
            "zona": zona,
            "tipo_evento": tipo_evento,
            "evento_activo": evento_activo,
            "asistentes_estimados": asistentes_estimados,
            "impacto_trafico_pct": impacto_trafico
        })

df_eventos = pd.DataFrame(filas)
print(df_eventos.shape)
df_eventos.head(10)
```

```python
df_eventos.to_csv("eventos_quito_simulado.csv", index=False)
```
